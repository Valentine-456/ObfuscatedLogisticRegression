# Data preprocessing and EDA of Bank Marketing Data Set

**The dataset is taken from here: [https://www.kaggle.com/datasets/ruthgn/bank-marketing-data-set/data](https://www.kaggle.com/datasets/ruthgn/bank-marketing-data-set/data)**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
```

```python
dataset = pd.read_csv('./data/bank-direct-marketing-campaigns.csv')
```

```python
dataset
```

```python
dataset.describe()
```

```python
dataset.shape, dataset.columns
```

```python
dataset = dataset.drop(columns=['contact', 'day_of_week', 'default', 'previous'])
```

```python
dataset
```

```python
print(dataset.isnull().sum())
```

```python
dataset = dataset[~dataset.isin(['unknown']).any(axis=1)]
```

```python
dataset
```

```python
cat_cols = dataset.select_dtypes(include='object').columns
for col in cat_cols:
    print(f"\n{col} — {dataset[col].nunique()} unique values:")
    print(dataset[col].value_counts())
```

```python
mappings = {}
for col in dataset.select_dtypes(include='object').columns:
    unique_vals = dataset[col].unique()
    mapping = {val: i for i, val in enumerate(sorted(unique_vals))}
    print(f"{col}_map = {mapping}")
    mappings[col] = mapping
```

```python
for col, mapping in mappings.items():
    dataset[col] = dataset[col].map(mapping)

# print("Unmapped values (NaN count) per column:")
# print(dataset[list(mappings.keys())].isnull().sum())

print("\nSample after encoding:")
print(dataset.head())
```

```python
dataset
```

```python
scaler = MinMaxScaler()
cols_to_scale = [col for col in dataset.columns if col != 'y']
dataset[cols_to_scale] = scaler.fit_transform(dataset[cols_to_scale])

dataset
```
